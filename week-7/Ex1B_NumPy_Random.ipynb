# Exercise 1.B NumPy: Random Numbers

```python
import numpy as np
```

## Generating Random Numbers

```python
r_1 = np.random.randint(1, 21, size=2)
print(r_1)

r_2 = np.random.randint(1, 21, size=(1, 2))
print(r_2)

r_3 = np.random.rand(2, 1)
print(r_3)

r_4 = np.random.randint(20, 101, size=(4, 5))
print(r_4)

r_5 = np.random.rand(4, 5)
print(r_5)
```

## Generating Random Choices from an Array

```python
students = np.array([
    "Hannah", "Sally", "Luis", "Alexis",
    "Corey", "Blake", "Leslie", "Livrata",
    "Charles", "Grace", "Ken", "Dagmawi"
])

call_on = np.random.choice(students)
print(call_on)

breakout_groups = np.random.choice(students, size=(3, 4))
print(breakout_groups)

# This may not be the best way to assign breakout groups because the same student can be selected more than once.
```

```python
call_on_weighted = np.random.choice(
    students,
    size=5,
    p=[0.30, 0.15, 0.10, 0.10, 0.08, 0.07, 0.06, 0.05, 0.04, 0.03, 0.02, 0.00]
)

print(call_on_weighted)

# Hannah has a 30% chance of being selected, Dagmawi has a 0% chance, and the other students have smaller probability values.
```

## Generating a Normal Distribution

```python
bell_curve = np.random.normal(loc=10, scale=5, size=(4, 5))
print(bell_curve)
```
