# Exercise 2.C Working with multiple DataFrames

```python
import pandas as pd
```

```python
employees = {
    "EmployeeID": [1, 2, 3, 4],
    "Name": ["Alice", "Bob", "Charlie", "Diana"],
    "DepartmentID": [101, 102, 101, 103]
}

departments = {
    "DepartmentID": [101, 102, 103],
    "DepartmentName": ["HR", "Finance", "IT"]
}

salaries = {
    "EmployeeID": [1, 2, 3, 4],
    "Salary": [60000, 75000, 50000, 90000]
}

df_employees = pd.DataFrame(employees)
df_departments = pd.DataFrame(departments)
df_salaries = pd.DataFrame(salaries)

df_employees
df_departments
df_salaries
```

```python
merged_df = pd.merge(df_employees, df_departments, on="DepartmentID")
merged_df
```

```python
final_df = pd.merge(merged_df, df_salaries, on="EmployeeID")
final_df
```

```python
high_salary = final_df[final_df["Salary"] > 60000]
high_salary
```

```python
grouped = final_df.groupby("DepartmentName")["Salary"].mean()
grouped
```

The merge function combines related data from different DataFrames using matching columns. Groupby helps summarize data by category, such as finding the average salary for each department.
