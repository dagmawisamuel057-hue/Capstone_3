# Exercise 5A — NumPy Fundamentals

This notebook practices the basics of NumPy arrays, including creating arrays, checking data types, transposing arrays, slicing values, and creating a 3D array.

## Array Basics

```python
fruits = ["apple", "banana", "cherry", "date"]
print(fruits)
```

```python
arr_1 = np.array(fruits)
print(arr_1)
```

```python
print(arr_1.dtype)

# U means Unicode string
# Number shows max character length
```

## Transposing Arrays

```python
arr_2 = np.array([
    [10, 11, 12],
    [13, 14, 15]
])

print(arr_2)
```

```python
print(arr_2.T)
```

```python
print(arr_2)

# arr_2.T does not permanently change arr_2
# it only displays a transposed version
```

## Array Slicing

```python
arr_3 = np.array([
    [1, 3, 5],
    [2, 4, 6]
])

print(arr_3)
```

```python
arr_4 = np.array([
    arr_3[0, 2],
    arr_3[1, 1]
])

print(arr_4)
```

## Three-Dimensional Arrays

```python
arr_5 = np.ones((2, 2, 2))

print(arr_5)
```
