# Exercise 1.A NumPy: Manipulating Arrays

```python
import numpy as np
```

```python
numbers = np.array([10, 20, 30, 40, 50])
print(numbers)
```

```python
print(numbers.shape)
```

```python
print(numbers.dtype)
```

```python
print(numbers[0])
print(numbers[-1])
print(numbers[1:4])
```

```python
arrA = np.array([[11, 12, 13], [14, 15, 16]])
print(arrA)

for x in arrA:
    print(x * 2)

for x in arrA:
    for y in x:
        print(y * 2)
```

```python
loop_result = []

for x in arrA:
    for y in x:
        loop_result.append(y * 2)

print(loop_result)
```

The first loop worked row by row, while the nested loop worked through each individual number inside the array.

```python
joined_array_1 = np.concatenate((arrA, arrA))
print(joined_array_1)

joined_array_2 = np.concatenate((arrA, arrA), axis=1)
print(joined_array_2)
```

```python
stacked_array = np.stack((arrA, arrA))
print(stacked_array)

horizontal_stack = np.hstack((arrA, arrA))
print(horizontal_stack)

vertical_stack = np.vstack((arrA, arrA))
print(vertical_stack)
```

```python
split_array = np.array_split(horizontal_stack, 2)
print(split_array)
```

```python
print(split_array[0])
print(split_array[1])
```

```python
sample = np.array([4, 9, 4, 8, 0, 6])

result = np.where(sample < 5)
print(result)

changed = np.where(sample < 5, sample, sample * 0.9)
print(changed)
```

```python
unsorted = np.array([14, 9, 4, 18, 0, 6])

print(np.sort(unsorted))

filtered = unsorted[unsorted > 10]
print(filtered)
```
