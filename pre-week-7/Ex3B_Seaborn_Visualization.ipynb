# Exercise 3.B Data Visualization with Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sales_data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [200, 250, 300, 280, 350, 400],
    "Profit": [50, 60, 75, 65, 90, 100]
}

df_sales = pd.DataFrame(sales_data)

df_sales
```

```python
sns.lineplot(x="Month", y="Sales", data=df_sales)
plt.title("Monthly Sales")
plt.show()
```

```python
sns.barplot(x="Month", y="Profit", data=df_sales)
plt.title("Monthly Profit")
plt.show()
```

```python
sns.scatterplot(x="Sales", y="Profit", data=df_sales)
plt.title("Sales vs Profit")
plt.show()
```

```python
sns.histplot(df_sales["Sales"])
plt.title("Sales Distribution")
plt.show()
```

Seaborn makes charts look cleaner and easier to customize. It works well with pandas DataFrames and is useful for professional-looking visualizations.
