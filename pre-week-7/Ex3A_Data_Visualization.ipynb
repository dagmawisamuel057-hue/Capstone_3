# Exercise 3.A Data Visualization with Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sales_data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [200, 250, 300, 280, 350, 400]
}

df_sales = pd.DataFrame(sales_data)

df_sales
```

```python
plt.plot(df_sales["Month"], df_sales["Sales"])
plt.show()
```

```python
plt.plot(df_sales["Month"], df_sales["Sales"], marker="o")
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
plt.bar(df_sales["Month"], df_sales["Sales"])
plt.title("Monthly Sales Bar Chart")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
plt.scatter(df_sales["Month"], df_sales["Sales"])
plt.title("Monthly Sales Scatter Plot")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

Line charts show trends over time, bar charts compare values between categories, and scatter plots help visualize relationships between data points.
