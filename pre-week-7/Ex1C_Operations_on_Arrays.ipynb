## Sample Arrays

```python
r_1 = np.random.randint(1, 20, size=2)
r_2 = np.random.randint(1, 20, size=(1, 2))
r_3 = np.random.rand(2, 1)
r_4 = np.random.randint(20, 100, size=(4, 5))
r_5 = np.random.rand(4, 5)

print(r_1)
print(r_2)
print(r_3)
print(r_4)
print(r_5)
```

## Permutations

```python
print(r_4)

np.random.shuffle(r_4)
print(r_4)
```

The shuffle function changed the order of the rows in r_4 directly. It modifies the original array in place.

```python
new_shuffle = np.random.permutation(r_4)

print(new_shuffle)
print(r_4)
```

I would use shuffle when I want to change the original array. I would use permutation when I want a new shuffled copy while keeping the original array unchanged.

```python
print(r_2)
print(r_2.shape)

print(r_4)
print(r_4.shape)

result = np.add(r_4, np.array([1, 1, 1, 1, 1]))
print(result)
print(result.shape)
```

# Adding r_1 to r_4 gives an error because their shapes do not match for broadcasting.
# r_2 can be added to r_4 because r_2 has 1 row and 2 columns, but r_4 has 4 rows and 5 columns, so this may also create a shape error.
# r_3 can be added to r_4 because r_3 has 2 rows and 1 column, but r_4 has 4 rows and 5 columns, so this also may not broadcast correctly.

print(np.add(r_1, r_4))

```python
print(r_4 + np.array([1]))
```

```python
print(r_4 + np.array([[1]]))
```

```python
print(r_4 + np.array([1, 1, 1, 1, 1]))
```

# This creates an error because the shape (2,1) cannot broadcast correctly with r_4 shape (4,5).
print(r_4 + np.array([[1], [1]]))

Broadcasting allows NumPy to combine arrays with different shapes when their dimensions are compatible. Smaller arrays can be stretched across larger arrays during the operation. The resulting shape usually follows the larger array when broadcasting is possible.

## Grocery Problem

```python
# Order: vegetables, fruits, snacks, beverages

bag1 = np.array([2, 3, 4, 0])
bag2 = np.array([4, 0, 1, 2])

print(bag1)
print(bag2)
```

```python
bag1_total = np.sum(bag1)
bag2_total = np.sum(bag2)

print(f"The first bag contains {bag1_total} items, and the second bag contains {bag2_total} items.")
```

```python
category_totals = bag1 + bag2

print(f"There are {category_totals[0]} vegetables, {category_totals[1]} fruits, {category_totals[2]} boxes of snacks, and {category_totals[3]} beverages in the two grocery bags.")
```

## Teamwork

```python
ali = np.array([5, 4, 6, 4, 2])
ben = np.array([0, 5, 5, 3, 4])
carla = np.array([3, 6, 0, 3, 4])

print(ali)
print(ben)
print(carla)
```

```python
all_knocks = np.array([ali, ben, carla])
print(all_knocks)
```

```python
person_totals = np.sum(all_knocks, axis=1)
group_total = np.sum(all_knocks)

print(f"Ali knocked on {person_totals[0]} doors.")
print(f"Ben knocked on {person_totals[1]} doors.")
print(f"Carla knocked on {person_totals[2]} doors.")
print(f"The group knocked on {group_total} doors total.")
```

```python
daily_totals = np.sum(all_knocks, axis=0)
rolling_total = np.cumsum(daily_totals)
estimated_votes = rolling_total * 0.5

print(estimated_votes)
```
